## Sales Trend Visualization

#### 1.Import Libraries

```python
import pandas as pd
import numpy as np  
import seaborn as sns
import matplotlib.pyplot as plt 
```

#### 2.Load Dataset

```python
df = pd.read_csv('../Data/sales_trend_data.csv')
df.head()
```

#### 3.Understand the data

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

#### 4.Data Cleaning

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df['Month'] = df['Date'].dt.month_name() # Month column is created..
print(df.head())
```

#### 5.Total Sales Analysis

```python
total_sales = df['Revenue'].sum()
print("Total Revenue:", total_sales)
```

#### 6.Monthly Sales Trend

```python
monthly_sales = df.groupby('Month')['Revenue'].sum()
print(monthly_sales)
```

```python
plt.figure(figsize=(10,6))
monthly_sales.plot(marker='o', linestyle='-', color='y')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

#### 7.Category-wise Sales

```python
category_sales = df.groupby('Category')['Revenue'].sum()
print(category_sales)
```

```python
category_sales.plot(kind='bar', color='c')
plt.title('Sales by Category')
plt.show()
```

#### 8.Top 10 Products

```python
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
top_products.plot(kind='bar', color='m')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```

#### 9.Key Insights

* Key Insights:

    1.Technology category generated the highest sales.

    2.West region contributed tha largest revenue.

    3.Sales increased significantly in the last quarter.

    4.Top 10 products contributed a major share of total revenue.

    5.Profit margins were highest in the Clothing category.
